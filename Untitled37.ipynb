```python
df['Watchtime in Million'] = df['Watchtime in Million'].str.replace('M', '').astype(float)
df['Watchtime'] = df['Watchtime'].str.replace(',', '').astype(float)
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Premiere'], kde=True)
plt.title('Distribution of Premiere')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Genre'], kde=True)
plt.title('Distribution of Genre')
plt.show()
```

```python
median_value = df['Premiere'].median()
df['Premiere'].fillna(median_value, inplace=True)
mode_value = df['Genre'].mode()[0]
df['Genre'].fillna(mode_value, inplace = True)
df.isnull().sum()
```

```python
release_counts = df['Premiere'].value_counts().sort_index()

# Plotting a bar chart with numbers on top of bars
plt.figure(figsize=(12, 6))
bars = release_counts.plot(kind='bar', color='skyblue')

# Adding numbers on top of the bars
for bar in bars.patches:
    plt.annotate(str(bar.get_height()),
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='bottom')

plt.title('Number of Releases for Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
unique_values = df['Premiere'].unique()

print(unique_values)
```

```python
# Extract decades
df['Decade'] = (df['Premiere'] // 10) * 10

# Plotting the decade-wise bar chart
plt.figure(figsize=(12, 6))
decade_counts = df['Decade'].value_counts().sort_index()
bars = decade_counts.plot(kind='bar', color='skyblue')

# Adding numbers on top of the bars
for bar in bars.patches:
    plt.annotate(str(bar.get_height()),
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='bottom')

plt.title('Number of Releases for Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Iterate through decades and plot bar graphs
for decade in df['Decade'].unique():
    subset = df[df['Decade'] == decade]
    release_counts = subset['Premiere'].value_counts().sort_index()

    # Plotting a bar chart with numbers on top of bars
    plt.figure(figsize=(12, 6))
    bars = release_counts.plot(kind='bar', color='skyblue')

    # Adding numbers on top of the bars
    for bar in bars.patches:
        plt.annotate(str(bar.get_height()),
                     (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                     ha='center', va='bottom')

    plt.title(f'Number of Releases for Each Year in Decade {decade}s')
    plt.xlabel('Year')
    plt.ylabel('Number of Releases')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.show()
```

```python
As we can see that there is a gradual increase in the number of releases each decade. Now we can check the changes in the genre of shows released over the years.
```

```python
# Group by 'Decade' and 'Genre' and count the number of releases
genre_counts_decade = df.groupby(['Decade', 'Genre']).size().unstack(fill_value=0)

# Plotting a stacked bar chart
plt.figure(figsize=(15, 8))
genre_counts_decade.plot(kind='bar', stacked=True)

plt.title('Number of Releases in Each Genre for Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Releases')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Group by 'Decade', 'Year', and 'Genre' and count the number of releases
genre_counts_decade = df.groupby(['Decade', 'Premiere', 'Genre']).size().unstack(fill_value=0)

# Iterate through decades and plot individual bar charts for each decade
for decade in df['Decade'].unique():
    subset = genre_counts_decade.loc[decade]

    # Plotting individual bar charts for each decade
    plt.figure(figsize=(15, 8))
    subset.plot(kind='bar', stacked=True)

    plt.title(f'Number of Releases in Each Genre for Decade {decade}s')
    plt.xlabel('Year')
    plt.ylabel('Number of Releases')
    plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
# Group by 'Year' and 'Genre' and count the number of releases
genre_counts_year = df.groupby(['Premiere', 'Genre']).size().unstack(fill_value=0)

# Iterate through genres and plot line charts
for genre in df['Genre'].unique():
    subset = genre_counts_year[genre]

    # Plotting a line chart
    plt.figure(figsize=(15, 8))
    plt.plot(subset.index, subset.values, label=genre)

    plt.title(f'Number of Releases for {genre} Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Number of Releases')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
# Create a violin plot
plt.figure(figsize=(15, 8))
sns.violinplot(x='Genre', y='Watchtime', hue='Genre', data=df, palette='muted', legend=False)

plt.title('Distribution of Watchtime for Different Genres')
plt.xlabel('Genre')
plt.ylabel('Watchtime')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import scipy.stats as stats

# Example correlation matrix
correlation_matrix = {
    'Rank': [1.000000, -0.133626, -0.386787, -0.147010],
    'Premiere': [-0.133626, 1.000000, 0.074435, 0.946622],
    'Watchtime': [-0.386787, 0.074435, 1.000000, 0.089211],
    'Decade': [-0.147010, 0.946622, 0.089211, 1.000000]
}

# Conduct hypothesis tests
alpha = 0.05
for variable, correlations in correlation_matrix.items():
    print(f'Hypothesis test for {variable}:')
    for other_variable, correlation_coefficient in correlation_matrix.items():
        if variable != other_variable:
            statistic, p_value = stats.pearsonr(correlations, correlation_coefficient)
            print(f'Correlation with {other_variable}: statistic={statistic}, p-value={p_value}')
            if p_value < alpha:
                print('Reject the null hypothesis (significant correlation)')
            else:
                print('Fail to reject the null hypothesis')
    print()
```

```python
These results show that the only relation is between Premiere and decade but that is not very essential relation for out analysis.
```
